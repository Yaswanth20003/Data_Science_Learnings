```python
import plotly.express as px
import pandas as pd
```

```python
df_cs=pd.read_csv('customer_segmentation_data.csv')
df_cs
```

```python
df_cs.info()
```

```python
df_cs['gender']=df_cs['gender'].astype('category')
df_cs['preferred_category']=df_cs['preferred_category'].astype('category')
```

```python
fig = px.box(df_cs,x='gender',y='spending_score',color='gender')
fig.show()
```

```python
fig=px.pie(df_cs, names='gender',values='income')
fig.show()
```

```python
fig = px.scatter(df_cs, x='income', y='spending_score', color='gender', title='Income and their Spending Score by Gender')
fig.show()
```

```python
fig=px.histogram(df_cs, x="membership_years",y='purchase_frequency',facet_col='gender', color='gender',nbins=20)
fig.show()
```

```python
fig=px.bar(df_cs, x='preferred_category',y='income')
fig.show()
```

```python
fig=px.line(df_cs,x='income',y='spending_score')
fig.show()
```

```python
fig=px.bar(df_cs,x='preferred_category',y='spending_score')
fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_cols=df_cs.select_dtypes(include=['float64', 'int64'])

sns.heatmap(numeric_cols.corr(), annot=True)
plt.show()
```
